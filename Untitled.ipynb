```python
import matplotlib.pyplot as plt
import re
import csv
```

```python
rawData = {}
with open('./startup.txt','r') as file:
    lines = csv.reader(file)
    num = 0
    for line in lines:
        if (len(line) == 0):
            continue
#         print(line)
        if ("processor" in line[0]):
            num = int(re.findall(r'\d+', line[0])[0])
            rawData.setdefault(num, [])
        else:
            res = float(re.findall("\d+\.\d+", line[0])[0])
            rawData[num].append(res)

print(rawData)
```

```python
keys = rawData.keys()
for i in keys:
    val = rawData[i]
    res = 0.0
    for j in val:
        if (res < j):
            res = j
    rawData[i] = res
    
print(rawData)
```

```python
x = list(rawData.keys())
y = list(rawData.values())
print(x)
print(y)

plt.style.use('ggplot')

plt.figure(figsize=(20,8),dpi=80)
plt.plot(x,y,label='start up time')
plt.xlabel('processor number')
plt.ylabel('time (s)')
plt.legend()
plt.show()
```
