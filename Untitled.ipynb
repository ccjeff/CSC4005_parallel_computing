```python
import matplotlib.pyplot as plt
import re
import csv
```

```python
rawData = {}
with open('./startup.txt','r') as file:
    lines = csv.reader(file)
    num = 0
    for line in lines:
        if (len(line) == 0):
            continue
#         print(line)
        if ("processor" in line[0]):
            num = int(re.findall(r'\d+', line[0])[0])
            rawData.setdefault(num, [])
        else:
            res = float(re.findall("\d+\.\d+", line[0])[0])
            rawData[num].append(res)

print(rawData)
```

```python
keys = rawData.keys()
for i in keys:
    val = rawData[i]
    res = 0.0
    for j in val:
        if (res < j):
            res = j
    rawData[i] = res
    
print(rawData)
```

```python
x = list(rawData.keys())
y = list(rawData.values())
print(x)
print(y)

plt.style.use('ggplot')

plt.figure(figsize=(20,8),dpi=80)
plt.plot(x,y,label='start up time')
plt.xlabel('processor number')
plt.ylabel('time (s)')
plt.legend()
plt.show()
```

```python
commData = {}
with open('./comm.txt','r') as file:
    lines = csv.reader(file)
    num = 0
    comm_len = 0
    key = 0
    pList = [2,4,8,16,20]
    for line in lines:
        if (len(line) == 0):
            continue
#         print(line)
        if ("processor" in line[0]):
            arr = re.findall(r'\d+', line[0])
            num = int(arr[0])
            if (num not in pList):
                continue
#             print(arr)
            comm_len = int(arr[1])
            key = str(num) + '_' + str(comm_len)
            commData.setdefault(key, [])
        else:
            res = float(re.findall("\d+\.\d+", line[0])[0])
            commData[key].append(res)

print(commData)
```

```python
keys = commData.keys()
plot_2 = []
plot_4 = []
plot_8 = []
plot_16 = []
plot_20 = []

for i in keys:
    val = commData[i]
    res = 0.0
    for j in val:
        res = res + float(j)
    np = int(i.split('_')[0])
#     print(np)
    commData[i] = res / np
    if (np == 2):
        plot_2.append(res / np)
    if (np == 4):
        plot_4.append(res / np)
    if (np == 8):
        plot_8.append(res / np)
    if (np == 16):
        plot_16.append(res / np)
    if (np == 20):
        plot_20.append(res / np)
    
    
print(commData)
```

```python
x = [0,1000,2000,3000,4000,5000,6000,7000,8000,9000]
plt.figure(figsize=(20,8),dpi=80)
plt.plot(x,plot_2[0:10],label='2 cores')
plt.plot(x,plot_4[0:10],label='4 cores')
plt.plot(x,plot_8[0:10],label='6 cores')
plt.plot(x,plot_16[0:10],label='10 cores')
plt.plot(x,plot_20[0:10],label='20 cores')


plt.xlabel('element length')
plt.ylabel('time (s)')
plt.legend()
plt.show()
```
